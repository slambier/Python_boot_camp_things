## Plotting the Ecliptic
The ecliptic, or the path that the Sun traces across the sky, is a simple sinusoidal curve. It's formula is as follows:

$ dec = i_E\sin(ra)$

where $dec$ is the declination, $i_E$ is Earth's inclination, and $ra$ is the right acension. The ecliptic is an important part to include when calculating the visibility of a sun-synchronous satellite:

```python
import numpy as np
import matplotlib.pyplot as plt

ra = np.linspace(0, 2*np.pi, 2000)
iearth_deg = 23.44
dec = iearth_deg*np.sin(ra)

fig = plt.figure(figsize=(14, 10))
ax = plt.gca()

plt.plot(ra/(2*np.pi)*360, dec, lw=5, color = "darkorchid")

plt.xlim(0, 360)
plt.ylim(-90, 90)

# Set axes colours
ax.tick_params(which="major", size=9, labelsize=16, width=1.5, direction='in')
#ax.tick_params(which="minor", size=5, labelsize=0, width=1.2, direction='in')

# Axes widths
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(1.5)

plt.xlabel("Right Ascension", fontsize = 18)
plt.ylabel("Declination", fontsize = 18)
plt.title("The Ecliptic", fontsize = 24)
```
